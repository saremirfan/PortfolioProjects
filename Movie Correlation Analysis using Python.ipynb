```python
# First we will import Libraries we are using in this Data Science Analysis Project.

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # This adjusts the configuration of the plots we will be creating.

# Reads in the data from a file from my downloads.

df = pd.read_csv(r'C:\Users\sarem\Downloads\movies.csv')
```

```python
# Let's look at the Movie Data taken from Kaggle.

df.head() # In the entire dataset, I use the df.head() command as it saves space when downloading.

# If you want to view the whole dataset, you could just set command as 'df'.
```

```python
# With these commands I will see if there is any missing data with a for loop.
# This will give us the percent of data that is missing per section.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing*100))
    
# This signifies to me that the data on Kaggle was not cleaned properly.
# I will still use this dataset for data analysis and understanding how the Python packeages work with visualization.
# In a future project, I will take the time to clean the data in the Excel and redo this project with a different dataset.
```

```python
# Similar to the last set of commands, with these I will see how many data points are missing exactly. 
# The data that comes up as "true" is in direct correlation with the percentage of data missing above.
# Again, I used a for loop to find this data.

for col in df.columns:
    print(df[col].isnull().value_counts(), "\n")
```

```python
# Data types for our columns.

df.dtypes
```

```python
# I will see if there are any outliers that exist in the data.

# Filtering the DataFrame to include only values above zero.

df_filtered = df[df['gross'] > 0]

# Creates a box plot for the filtered data.

plt.boxplot(df_filtered['gross'])
plt.show()

# As we can see there do happen to be some outliers that do exist in the movies.
```

```python
# With these commands I am changing the data types of some of the columns to integers.

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# df
```

```python
# Creating a correct Year column.
# This is because year and released columns don't match exactly for some reason.

df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

df.head()
```

```python
# Ordering the data by gross revenue.

# Sort the DataFrame by 'gross' column in descending order and get the top 'n' rows (e.g., top 5 rows).

top_n = 5  # You can change this value to the number of top rows you want to see.
sorted_df = df.sort_values(by='gross', ascending=False).head(top_n)

# Display the sorted and filtered DataFrame.
print(sorted_df)

# As we can see based on this, Avatar, Avengers:Endgame, and Titanic were the top three grossing films.
```

```python
pd.set_option('display.max_rows', None) # This is setting max rows for the last line.
```

```python
# Dropping any duplicates with this function.

df_no_duplicates = df.drop_duplicates()

# Use the head method to display only the first few rows (e.g., the first 5 rows).
df_no_duplicates_head = df_no_duplicates.head()

# Display the resulting DataFrame.
print(df_no_duplicates_head)

# This has now dropped any duplicate data points that exist.
```

```python
# Scatter plot with budget vs gross using Seaborn.

# Filter the DataFrame to include only data points in the top right quadrant as we have outliers.
top_right_quadrant_df = df[(df['gross'] > 0) & (df['budget'] > 0)]

# Create a scatterplot for the filtered data.
sns.regplot(x="gross", y="budget", data=top_right_quadrant_df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})

# Set labels and titles for the plot.
plt.xlabel("Budget")
plt.ylabel("Gross Earnings")
plt.title("Regression Plot for Budget vs Gross Earnings")

# Show the plot.
plt.show()
```

```python
# Scatter plot with score and gross using Seaborn.

# Filter the DataFrame to include only data points in the top right quadrant as we have outliers.
top_right_quadrant_df = df[(df['score'] > 0) & (df['gross'] > 0)]

# Create a scatterplot for the filtered data.
sns.regplot(x="score", y="gross", data=top_right_quadrant_df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})

# Set labels and titles for the plot.
plt.xlabel("Score")
plt.ylabel("Gross")
plt.title("Regression Plot for Score and Gross")

# Show the plot.
plt.show()
```

```python
# Let's start looking at correlation.
# As we can see, budget and gross have a high correlation below.

df.corr(numeric_only=True) # This is using the Pearson Model as default. 

# Although I can use Kendall and Spearman if I wanted.
```

```python
# Using the correlation matrix to further analyze.

correlation_matrix = df.corr(numeric_only=True) # This is using the Pearson Model as default. 

#Although I can use Kendall and Spearman if I wanted.

sns.heatmap(correlation_matrix, annot=True)

# Set labels and titles for the plot.
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.title("Correlation Matrix for Numeric Features")

plt.show()

# Below we see that the numbers that are closer to 1 have a higher correlation.
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
# Using the correlation matrix to further analyze.

correlation_matrix = df_numerized.corr(numeric_only=True) # This is using the Pearson Model as default. 

# Although I can use Kendall and Spearman if I wanted.

sns.heatmap(correlation_matrix, annot=True)

# Set labels and titles for the plot.
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.title("Correlation Matrix for Numeric Features")

plt.show()

# Below we see that the numbers that are closer to 1 have a higher correlation.
# This is showing us a correlation between each and every variable available.
```

```python
# With this we can see that gross is highly correlated with votes and budget in a different format.

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Here I will sort the correlated pairs.

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Using this we can now take a look at the ones that high a high correlation (>0.5).

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
